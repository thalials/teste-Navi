### Questão 1            

```python
def findNumbers(a, b):
    # numeros pares 
    lista = []
    numbers = []
    for number in range(a,b):
        if number % 2 == 0:
            lista.append(number)
            
    for n in lista:
        if n%49==0 and n%37==0:
            numbers.append(n)
            
    return numbers
```

```python
findNumbers(1, 5000000)
```

### Questão 2

```python
import math
import numpy as np
```

```python
def findVector(positions):
    # values of vector equals to 0
    X = [0]*positions
    
    i = 0 
    while i < len(X):
        if i % 2 == 0:
            X[i] = 3**i + 7*(math.factorial(i))
        else:
            X[i] = 2**i + 4*(np.log(i))
        i +=1 
        
    for i in [i for i,x in enumerate(X) if x == max(X)]:
        max_position = i
        
    media = np.mean(X)
    resultado = "A média dos elementos desse vetor é igual a {0:.2f}. A posição de valor máximo desse vetor é igual a {1}".format(media, max_position)
    return resultado
```

```python
findVector(10)
```

### Questão 3

```python
i = 0
student_scored = {}
listaScores = []

while i < 5:
    student = input("Qual o nome do aluno? ")
    scored = float(input("Qual a nota dele? "))
    
    if student not in student_scored:
        student_scored[student] = scored
        
    if scored not in listaScores:
        listaScores.append(scored)
        
    i += 1

scored_max = max(listaScores)

for student in student_scored:
    if student_scored[student] == scored_max:
        name = student

print("A maior nota foi {0} do aluno {1}".format(scored_max, name))
    
```
